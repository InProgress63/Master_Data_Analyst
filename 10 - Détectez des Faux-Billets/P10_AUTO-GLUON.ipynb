# Récupération des données :

```python
Billets = pd.read_csv("billets - billets.csv",sep=';')
Billets.head()
```

# Préparation des données :

```python
# Contrôle des Nans :
(Billets.isna().sum()/Billets.shape[0]).sort_values(ascending=True)
```

```python
# Billets = Billets.dropna()
```

```python
# Retrait des Nans :

# FillNa() pour les catégories par moyenne de représentabilité des séquences :
cats = Billets.columns

for i in Billets.columns:
    if Billets[i].isna().sum()>0:
        if i in cats:
            Billets[i]=Billets[i].fillna(Billets[i].mean())
```

```python
(Billets.isna().sum()/Billets.shape[0]).sort_values(ascending=True)
```

```python
# Imports librairies & packages :
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler
```

```python
X = Billets.drop(columns=['is_genuine'])
y = Billets['is_genuine']


X = X.sort_index()
y = y.sort_index()
```

```python
# TrainTestSplit :

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```
